```python
from ultralytics import YOLO
import cv2
import os
from pathlib import Path
import matplotlib.pyplot as plt # Import matplotlib

# ëª¨ë¸ ë¡œë“œ
model = YOLO("/content/runs/detect/human_detection_v84/weights/best.pt")  # í•™ìŠµëœ ëª¨ë¸ ê²½ë¡œ

# ì´ë¯¸ì§€ í´ë” ê²½ë¡œ
image_folder = "/content/drive/MyDrive/capstone/HumanDetection/predict"
output_folder = "/content/drive/MyDrive/capstone/HumanDetection/output"

# ì¶œë ¥ í´ë”ê°€ ì—†ìœ¼ë©´ ìƒì„±
os.makedirs(output_folder, exist_ok=True)

# í´ë” ë‚´ ëª¨ë“  ì´ë¯¸ì§€ íŒŒì¼ ë¦¬ìŠ¤íŠ¸ (í™•ì¥ì í•„í„°ë§)
image_paths = [p for p in Path(image_folder).glob('*') if p.suffix.lower() in ['.jpg', '.jpeg', '.png']]

for img_path in image_paths:
    # ì´ë¯¸ì§€ ì½ê¸°
    img = cv2.imread(str(img_path))
    # OpenCVëŠ” BGRì´ê¸° ë•Œë¬¸ì— RGBë¡œ ë³€í™˜
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # ì¶”ë¡  ìˆ˜í–‰
    results = model(img_rgb)

    # ê²°ê³¼ ì‹œê°í™”
    # results.show()ëŠ” OpenCV ìœˆë„ìš° ë˜ëŠ” Colabì—ì„œ ë°”ë¡œ ë³´ì—¬ì¤„ ìˆ˜ ìˆìŒ
    # access the first element of the results list
    results[0].show()

    # ë˜ëŠ” Matplotlibìœ¼ë¡œ ë³´ì—¬ì£¼ê¸°
    # results.render()ëŠ” ë°”ìš´ë”© ë°•ìŠ¤ì™€ ë¼ë²¨ì´ ê·¸ë ¤ì§„ ì´ë¯¸ì§€ë¥¼ ë°˜í™˜
    rendered_img = results[0].render()[0]  # ë¦¬ìŠ¤íŠ¸ì—ì„œ ì²« ë²ˆì§¸ ì´ë¯¸ì§€
    plt.figure(figsize=(8,8))
    plt.imshow(rendered_img)
    plt.axis('off')
    plt.show()

print(f"ì¶”ë¡  ì™„ë£Œ. ê²°ê³¼ëŠ” {output_folder} í´ë”ì— ì €ì¥ë¨.")
```
