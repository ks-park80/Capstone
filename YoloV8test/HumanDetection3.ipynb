```python
from ultralytics import YOLO
import cv2
import os
from pathlib import Path
import matplotlib.pyplot as plt # Import matplotlib

# 모델 로드
model = YOLO("/content/runs/detect/human_detection_v84/weights/best.pt")  # 학습된 모델 경로

# 이미지 폴더 경로
image_folder = "/content/drive/MyDrive/capstone/HumanDetection/predict"
output_folder = "/content/drive/MyDrive/capstone/HumanDetection/output"

# 출력 폴더가 없으면 생성
os.makedirs(output_folder, exist_ok=True)

# 폴더 내 모든 이미지 파일 리스트 (확장자 필터링)
image_paths = [p for p in Path(image_folder).glob('*') if p.suffix.lower() in ['.jpg', '.jpeg', '.png']]

for img_path in image_paths:
    # 이미지 읽기
    img = cv2.imread(str(img_path))
    # OpenCV는 BGR이기 때문에 RGB로 변환
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # 추론 수행
    results = model(img_rgb)

    # 결과 시각화
    # results.show()는 OpenCV 윈도우 또는 Colab에서 바로 보여줄 수 있음
    # access the first element of the results list
    results[0].show()

    # 또는 Matplotlib으로 보여주기
    # results.render()는 바운딩 박스와 라벨이 그려진 이미지를 반환
    rendered_img = results[0].render()[0]  # 리스트에서 첫 번째 이미지
    plt.figure(figsize=(8,8))
    plt.imshow(rendered_img)
    plt.axis('off')
    plt.show()

print(f"추론 완료. 결과는 {output_folder} 폴더에 저장됨.")
```
